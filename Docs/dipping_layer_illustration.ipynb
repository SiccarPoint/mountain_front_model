# How to set parameters for a dipping layer of rock

```python
def draw_layer(theta, thickness, z0_top=0.0, layer_left_x=0.0, no_show=False):
    """Calculate and draw a dipping layer with specified params."""
    import numpy as np
    import matplotlib.pyplot as plt
    %matplotlib inline

    if theta == 90.0:  # special case
        z = np.arange(-thickness * 5, thickness * 5 + 1, thickness * 5)
        x_left = np.zeros(z.size) + layer_left_x
        x_right = x_left + thickness
        plt.plot(x_left, z)
        plt.plot(x_right, z)
        if not no_show:
            plt.show()
    else:
        x = np.arange(0.0, 10 * thickness, 0.1 * thickness)
        tantheta = np.tan(np.pi * theta / 180.0)
        z_top = tantheta * x + z0_top
        z_bottom = tantheta * x + (z0_top - thickness / np.cos(np.pi * theta / 180.0))
        plt.plot(x, z_top)
        plt.plot(x, z_bottom)
        if not no_show:
            plt.show()
```

```python
draw_layer(90.0, 1.0)
```

```python
draw_layer(45.0, 1.0)
```

```python
from landlab import HexModelGrid
import matplotlib.pyplot as plt
%matplotlib inline

hg = HexModelGrid(5, 5, orientation='vert', shape='rect')
draw_layer(45.0, 1.0, z0_top=-1.0, no_show=True)
plt.plot(hg.x_of_node, hg.y_of_node, '.')
plt.show()
```
